## Laboratorio 2 Git y Github
#### Introducción

Git: es un un software encargada del manejo de versiones confiando en la eficiencia del manejo de proyectos con gran cantidad de archivos, se trabaja en un repositorio donde se almacenan los archivos de nuestros proyectos existen dos tipos de repositorio el local y el remoto. Git es el software local y Github el remoto.

Se trabaja por ramas siendo un espacio del repositorio donde se almaceneran los archivos, a la rama principal se le llama origen master. Este se compone por varios comandos los cuales son:

##### Areas de trabajo
- Working Directory: trabajo los archivos enforma local
- Stagin área: al generar algún cambio y guardara en la nube la documentación, preparándola para su versión final, ya que los archivos están modificados se traslada a la siguiente área
- .git directory: área de commit o publicación final del proyecto o archivos
